```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
df_sorted = df.sort_values("area")

plt.plot(df_sorted["area"],df_sorted["price"], color = "blue")
plt.title("house Area v/S Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
furnish_counts = df["furnishingstatus"].value_counts()
furnish_counts

plt.pie(furnish_counts,labels=furnish_counts.index)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True , cmap="coolwarm")
plt.show()
```

```python
sns.boxplot(x="airconditioning", y="price", data=df)
plt.show()
```

```python
sns.histplot(df["price"], kde=True,)
plt.show()
```
